```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.scatter(df.area, df.price, color='red', marker='+')
plt.xlabel('area')
plt.ylabel('price(US$)')
```

```python
reg = linear_model.LinearRegression()
# Fit is used to train the data
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
# It is used to show the value of the slope
reg.coef_
```

```python
# It is used to show the value of the constant
reg.intercept_
```

```python
d = pd.read_csv('areas.csv')
d
```

```python
p = reg.predict(d)
d['prices'] = p
d.to_csv('prediction.csv', index=False)
```
